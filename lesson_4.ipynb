```python
# Вася и носки

n, m = map(int, input().split())
days = 0
while True:
    if n < m:
        days += n
        break
    else:
        days += (n // m * m)
        n = (n // m) + (n % m)
print(days)
```

```python
# Собираемся в поход

n = int(input())
count = 0
total = 0
while total <= n:
    thing = int(input())
    total += thing
    count += 1
print('Довольно!', total - thing, count - 1, sep='\n')
```

```python
# Новый год и спешка

n, k = map(int, input().split())
time = 240 - k
count = 0
time_task = 5
while time - time_task >= 0 and count < n:
    count += 1
    time -= time_task
    time_task += 5
print(count)
```

```python
# Ваня и кубики

n = int(input())
floor = num = 0
while n - num - floor - 1 >= 0:
    floor += 1
    num += floor
    n -= num
print(floor)
```

```python
# Слияние списков

n, m = map(int, input().split())
a, b = list(map(int, input().split())), list(map(int, input().split()))
res = []
while len(a) > 0 and len(b) > 0:
    if a[0] < b[0]:
        res.append(a[0])
        a = a[1:]
    else:
        res.append(b[0])
        b = b[1:]
res += a
res += b
print(*res)
```
